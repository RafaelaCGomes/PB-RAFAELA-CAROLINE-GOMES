```python
import pandas as pd

#abrir movies.csv

caminho_movies = 'C:/Users/User/Desktop/Compass.uol/Sprint 7/Desafio/Etapa-01/movies.csv'

df_movies = pd.read_csv(caminho_movies, delimiter='|')
df_movies.head()
```

```python
print(df_movies.info())
```

```python
df_fantasi = df_movies.loc[df_movies['genero'].str.contains('fantasy', case=False, na=False)  ]
df_fantasi
```

```python
df_star = df_movies.loc[df_movies['tituloPincipal'].str.contains('Star Wars', case=False, na=False)]
df_star
```

```python
df_star_decrescente = df_star.sort_values(by='notaMedia', ascending=False)
df_star_decrescente.head(20)
```

```python
df_fantasi['anoLancamento'] = pd.to_numeric(df_fantasi['anoLancamento'], errors='coerce')

# Filtrar os filmes do gênero fantasia lançados entre 1977
filmes_filtrados = df_fantasi.loc[(df_fantasi['anoLancamento'] >= 1970) & (df_fantasi['anoLancamento'] <= 2019)]
filmes_filtrados
```

```python
df_unicos = filmes_filtrados.drop_duplicates(subset='id')
df_unicos
```

```python
df_unicos.to_csv('filmes_unicos.csv', index=False)
print("O dataframe foi salvo como 'filmes_unicos.csv'")
```
