### Etapa 2 - Item 2: Criação de gráfico de barras contendo os top 5 apps por numero de instalação.

- Comecei realizando a conversão de dados da coluna Installs para o tipo numérico, pois a coluna estava sendo tratada como texto.
- Após a conversão, realizei a seleção dos 5 apps com maior número de instalaçõe utilizando a função `nlargest()`.

```python

ds_sem_dupli['Installs'] = pd.to_numeric(ds_sem_dupli['Installs'].str.replace(',', '').str.replace('+', ''), errors='coerce')

top_5_apps = ds_sem_dupli.nlargest(5, 'Installs')[['App', 'Installs']]
top_5_apps
```

- Criação do gráfico "Top 5 app mais instalados"

```python


#Dados do gráfico
Apps = top_5_apps['App']
Installs = top_5_apps['Installs']

plt.figure(figsize=(11, 9)) 

plt.bar(Apps,Installs, color='aquamarine', edgecolor='black')

plt.title ('Top 5 aplicativos mais instalados') #Titulo

plt.xlabel('Apps')
plt.ylabel('Installs')
plt.xticks(rotation=90)

plt.tight_layout() #ajusta a legenda para não sair da janela

plt.savefig('Top_5_app.jpg') #Exportação em jpg.

plt.show()
```

### Etapa 2 - Item 3: Criação do gráfico que contém as categorias de apps existentes no dataset, de acordo com a frequencia em que aparecem.  


- Comecei conferindo a frequencia de cada categoria


```python
freq_categoria = ds_sem_dupli['Category'].value_counts()
freq_categoria
```

- Criação do gráfico de pizza.

```python
# Adicionando um título

plt.figure(figsize=(11, 9)) 

plt.pie(freq_categoria,
        labels=freq_categoria.index, 
        autopct='( %1.1f%% )', 
        pctdistance=1.1,
        startangle=180, 
        labeldistance=1.30,
        textprops={'ha': 'center', 'va': 'center', 'fontsize': 6})

plt.title('Frequência de Categorias de Apps', loc='left', pad=44)
# Exibindo o gráfico
plt.axis('equal')  # Para garantir que o gráfico seja um círculo

plt.savefig('Frequencia_de_categorias_de_apps.jpg') #Exportação em jpg.

plt.show()
```

### Etapa 2 - Item 4 : Mostrar qual o app mais caro existente no DataSete.

- Comecei convertendo a coluna Price para numeric.
- Em seguida, usei a função `nlargest()` para encontrar o app mais caro.

```python
ds_sem_dupli['Price'] = pd.to_numeric(ds_sem_dupli['Price'].str.replace(',', '').str.replace('$', ''), errors='coerce')

app_mais_caro = ds_sem_dupli.nlargest(1, 'Price')[['App', 'Price']]
print('O aplicativo mais caro e:')
app_mais_caro
```

### Etapa 2 - Item 5: Mostrar quantos apps são classificados como 'Mature17+'

### Etapa 2 - Item 6: Mostrar o top 10 apps por numero de reviews bem como o respectivo número de reviews.
 
- Comecei convertendo a coluna "reviews" para numeric.
- Em seguida, usei a função `groupby()`para agrupar os dados por nome de app.
- Com a função `sum()`, somei os valores das Review de cada app.
- Em seguida, usei a função `sort_values()` e `asceding=False` para ordenar a coluns em ordem decrescente.

```python
ds_sem_dupli['Reviews'] = pd.to_numeric(ds_sem_dupli['Reviews'], errors='coerce')

grupo_apps = ds_sem_dupli.groupby('App', as_index=False)['Reviews'].max()


top_10_apps = grupo_apps.sort_values(by='Reviews', ascending=False).head(10)[['App', 'Reviews']]


print('O top 10 aplicativos com maior número de Reviews são:')
print(top_10_apps)
```

### Etapa 2 - Item 7: Criar dois cálculos sobre o dataset, um em formato de lista e outro em formato de valor.

- Formato de lista: Mostrar os 15 apps mais leves (menor size).

Comecei convertendo os dados da coluna 'Size' para numeric. Após isso, usei a função `sort_values()` para ordenar os dados em ordem crescente com `ascending=True`. Verificando assim os 20 apps mais leves.

```python
ds_sem_dupli['Size'] = pd.to_numeric(ds_sem_dupli['Size'].str.replace(',', '').str.replace('M', ''), errors='coerce')


mais_leves = ds_sem_dupli.sort_values(by='Size', ascending=True).head(15)[['App', 'Size']]
mais_leves
```

- Formato de valor: Mostrar o numero total de apps que são do gênero Finance.


```python
apps_Finance = ds_sem_dupli['Genres'].value_counts().get('Finance', 0)
print(f"Existe {apps_Finance} aplicativos do genero 'Finance' no dataset.")
```

### Etapa 2 - Item 8: Criar dois gráficos de exibição dos indicadores dos itens realizados.

- Gráfico do top 10 apps por numero de reviews bem como o respectivo número de reviews.

```python

#dados do gráfico: 
#Apps_10= top_10_apps['App']
#Reviews_10= top_10_apps['Reviews']

plt.figure(figsize=(11,8))
plt.plot(top_10_apps['App'], top_10_apps['Reviews'], marker='o', linestyle='-', color='red')


plt.title('Top 10 Aplicativos com Maior Número de avaliações (Review)')

plt.xlabel('Apps')
plt.xticks(rotation=90) 

plt.ylabel('Número de Reviews (Milhões)')
plt.yticks(np.arange(10000000,100000000, 10000000)) #Alt do intervalo de marcação do eixo y.



for i, row in top_10_apps.iterrows(): #Mostra o valor exato de review de cada app.
    plt.text(row['App'], row['Reviews'], f'({row["Reviews"]:,})', ha='center', va='bottom', fontsize=6, rotation=45)


plt.tight_layout() #ajusta a posição dos textos para não sair da figura.

plt.savefig('Top_10_apps_com_maior_avaliação.jpg') #Exportação em jpg.

plt.show()
```
